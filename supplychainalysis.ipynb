```python
import pandas as pd

df = pd.read_csv("Inventory_SupplyChain_Dataset.csv")
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df["Inventory Utilization %"] = (df["Units Sold"] / df["Inventory Level"]) * 100
```

```python
df["Warehouse Utilization %"] = (df["Inventory Level"] / df["Warehouse Capacity"]) * 100
```

```python
df["Lead Time Category"] = df["Lead Time (Days)"].apply(
    lambda x: "Fast" if x <= 3 else ("Medium" if x <= 7 else "Slow")
)
```

```python
df["Backorder Status"] = df["Backorder"].apply(
    lambda x: "Yes" if x == True else "No"
)
```

```python
df.to_csv("final_supply_chain.csv", index=False)
```
